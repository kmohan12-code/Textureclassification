```python
import timm
import torch

model = timm.create_model('vit_base_patch16_224', pretrained=False, num_classes=47)
```

```python

model.load_state_dict(torch.load('/content/drive/MyDrive/saved Model/best_model (1).pth', map_location=torch.device('cpu')))
model.eval()
```

```python
import torch
from PIL import Image
from torchvision import transforms

model.eval()

img = Image.open('/content/drive/MyDrive/dtd/images/blotchy/blotchy_0003.jpg')


transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
])

img = transform(img).unsqueeze(0)


classes = []
with open('/content/drive/MyDrive/saved Model/classnames.txt', 'r') as f:
    classes = [line.strip() for line in f]

with torch.no_grad():
    output = model(img)
    predicted_class = output.argmax(1)


predicted_label = classes[predicted_class.item()]
print(predicted_label)
```
